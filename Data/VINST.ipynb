```python
%%sql
DROP TABLE IF EXISTS Volvo CASCADE;
CREATE TABLE Volvo(
    SR_Number TEXT,
    Date_Time TIMESTAMP WITH TIME ZONE,
    Status TEXT,
    Sub_Status TEXT,
    Involved_ST_Function_Div TEXT,
    Involved_Org_line_3 TEXT,
    Involved_ST TEXT,
    SR_Latest_Impact TEXT,
    Product TEXT,
    Country TEXT,
    Owner_Country TEXT,
    Owner_First_Name TEXT);
```

```python
%%sql
SELECT *
FROM Volvo
LIMIT 10;
```

```python
%%sql
(SELECT sr_number, MAX(date_time) AS End_date
FROM Volvo
GROUP BY sr_number
ORDER BY End_date
LIMIT 1)

UNION

(SELECT sr_number, MAX(date_time) AS End_date
FROM Volvo
GROUP BY sr_number
ORDER BY End_date DESC
LIMIT 1);
```

```python
%%sql
WITH count AS (
SELECT DISTINCT sr_number, owner_country
FROM Volvo
)

SELECT sr_number, COUNT(*) AS cnt
FROM count
GROUP BY sr_number
ORDER BY sr_number DESC;
```
